```python
def checkpoint(file_name, checkpoint_header, checkpoint_data):
    np.savez(file_name, header = checkpoint_header, data = checkpoint_data)
    checkpoint_variable = np.load(file_name + ".npz")
    return(checkpoint_variable)
```

```python
df = raw_data.copy()
```

```python
df = df.drop(['ID'], axis = 1)
```

```python
# Define the bin edges for grouping
bins = [-1, 0, 14, 17, 21, 28]

# Define the labels for each group
labels = [0, 1, 2, 3, 4]

# Use the cut() function to assign groups based on the column values
df['Reason for Absence'] = pd.cut(df['Reason for Absence'], bins=bins, labels=labels, right = True)
```

```python
date_split = df['Date'].str.split('/', n = 2, expand = True)
date_split
```

```python
date_split = date_split.rename(columns = {0: 'Day', 1: 'Month', 2: 'Year'})
date_split
```

```python
date_split = date_split.drop(['Day','Year'], axis = 1)
```

```python
df['Date'] = pd.to_datetime(df['Date'], format = 'mixed')
```

```python
def date_to_weekday(date_value):
    return date_value.weekday()
```

```python
df['Day of the Week'] = df['Date'].apply(date_to_weekday)
```
